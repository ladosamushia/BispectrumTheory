```julia
using DelimitedFiles
using Statistics
using Plots
```

```julia
bktot = zeros(25,2600)
cc = 1
for ff in readdir("../data/")
    if ff[1:10] == "bk_LRGz0.8"
        fname = string("../data/", ff)
        bk = readdlm(fname)
        bktot[cc,:] = bk[:,4]
        cc += 1
    end
end
k123 = bk[:,1:3];
```

```julia
bkmean = transpose(mean(bktot, dims=1));
```

```julia
kbins = collect(0.0:0.01:0.3)
bk_k1 = average_over_k1(k123[:,1], kbins,bkmean);
```

```julia
scatter(kbins, bk_k1.*kbins)
```

```julia
writedlm("../data/bk_average_k1.txt", bk_k1)
```
